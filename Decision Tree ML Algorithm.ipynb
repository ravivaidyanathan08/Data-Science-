```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris.DESCR)
```

```python
print(iris.target)
```

```python
iris.columns
```

```python
X = pd.DataFrame(iris.data, columns=['sepal length in cm', 'sepal width in cm','petal length in cm','petal width in cm'])
```

```python
X
```

```python
y = iris.target
```

```python
y
```

```python
## Train, test split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
## Apply Decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
## Visualize the decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
plt.show()
```

```python
## Postpruning with a max depth equal to 2
## Apply Decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier(max_depth=2)
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
y_pred = treeclassifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

## Decision tree classifier prepruning

```python
param = {'criterion':['gini', 'entropy', 'log_loss'],'splitter':['best', 'random'],'max_depth':[1,2,3,4,5],
         'max_features':['sqrt', 'log2']}
```

param

## Decision tree regressor - Diabetes Dataset

```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
```

```python
print(dataset.DESCR)
```

```python
print(dataset.data)
```

```python
print(dataset.target)
```

```python
import pandas as pd
```

```python
df_diabetes = pd.DataFrame(dataset.data, columns = ['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
```

```python
X = df_diabetes
```

```python
y = dataset.target
```

```python
X
```

```python
y
```

```python
X.head()
```

```python
## Train, test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train.head()
```

```python
## Check correlation 
X_train.corr()
```

```python
## Visualize correlation
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(data=X_train.corr(),annot=True)
plt.show()
```

```python
## Apply the ML model
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train,y_train)
```

## Hyperparameter tuning

```python
param = {'criterion': ['squared_error','friedman_mse', 'absolute_error'],
         'splitter':['best','random'],
         'max_depth': [1,2,3,4,5,10,15,20,25],
         'max_features':['auto','sqrt', 'log2']
        }
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator =regressor,param_grid=param ,cv=5, scoring= 'neg_mean_squared_error')
```

```python
grid.fit(X_train,y_train)
```

```python
y_pred = grid.predict(X_test)
```

```python
grid.best_params_
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```
